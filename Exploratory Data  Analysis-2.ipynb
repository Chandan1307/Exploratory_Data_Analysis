## Q1. Load the flight price dataset and examine its dimensions. How many rows and columns does the dataset have?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings

warnings.filterwarnings("ignore")
```

```python
dataset = pd.read_excel('flight_price.xlsx')
```

```python
dataset
```

```python
num_rows, num_cols = dataset.shape

print(f"The dataset has {num_rows} rows and {num_cols} columns.")
```

## Q2. What is the distribution of flight prices in the dataset? Create a histogram to visualize the distribution.

```python
price_column_name = 'Price'

# Plot the histogram
plt.figure(figsize=(10, 6))
plt.hist(dataset[price_column_name], bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Flight Price')
plt.ylabel('Frequency')
plt.title('Distribution of Flight Prices')
plt.grid(True)
plt.show()
```

## Q3. What is the range of prices in the dataset? What is the minimum and maximum price?

```python
price_column_name = 'Price'

# Calculate the minimum and maximum prices
min_price = dataset[price_column_name].min()
max_price = dataset[price_column_name].max()
print(f"The minimum price is: {min_price}")
print(f"The maximum price is: {max_price}")
```

```python
# Calculate the range of prices
price_range = max_price - min_price
print(f"The range of prices in the dataset is: {price_range}")
```

## Q4. How does the price of flights vary by airline? Create a boxplot to compare the prices of different airlines.

```python
airline_column_name = 'Airline'
price_column_name = 'Price'
# Create a boxplot to compare prices by airline
plt.figure(figsize=(12, 8))
sns.boxplot(x=airline_column_name, y=price_column_name, data=dataset)
plt.xlabel('Airline')
plt.ylabel('Flight Price')
plt.title('Price of Flights by Airline')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.show()
```

## Q5. Are there any outliers in the dataset? Identify any potential outliers using a boxplot and describe how they may impact your analysis.

```python
price_column_name = 'Price'

# Create a boxplot to identify potential outliers
plt.figure(figsize=(8, 6))
plt.boxplot(dataset[price_column_name], vert=False)
plt.xlabel('Flight Price')
plt.title('Boxplot of outliers')
plt.grid(True)
plt.show()
```

## Q11. Are there any missing values in the dataset? Identify any missing values and describe how they may impact your analysis.

```python
missing_values = df.isnull().sum()

# Display columns with missing values and their corresponding counts
print(missing_values[missing_values > 0])
```

The impact of missing values on your analysis depends on the extent of missingness and the nature of the analysis you are performing:

Data Integrity: Missing values can lead to data integrity issues, especially if they are distributed across critical columns.

Biased Analysis: If the missing data is not missing at random, it can lead to biased analysis or incorrect conclusions.

Reduced Sample Size: If a significant number of observations have missing values, it may reduce the effective sample size for analysis.

Impact on Statistical Analysis: Missing values can affect statistical analysis, as some calculations may exclude missing values or impute them with a default value, leading to biased estimates.

Machine Learning Models: Some machine learning algorithms may not handle missing values well, and imputing missing data can introduce noise or affect model performance.

Imputation Strategies: Choosing an appropriate imputation strategy is crucial, as different methods can impact the data and analysis differently.

To mitigate the impact of missing values, you can consider the following strategies:

Data Imputation: Impute missing values using techniques like mean, median, mode, or more advanced methods like regression imputation.

Complete-Case Analysis: If the missingness is minimal, you can perform complete-case analysis by excluding observations with missing values.

Multiple Imputation: Use multiple imputation techniques to estimate missing values multiple times and create multiple datasets for robust analysis.

Domain Expertise: Consult domain experts to determine if the missing data can be reasonably estimated or if certain missing values should be excluded from the analysis.

Remember to document the handling of missing values in your analysis to ensure transparency and replicability of the results. Careful consideration and appropriate handling of missing data are essential for accurate and reliable analysis results.

## Q12. What is the relationship between the size of an app and its rating? Create a scatter plot to visualize the relationship.

```python
# Create a scatter plot to visualize the relationship between app size and rating
plt.figure(figsize=(10, 6))
plt.scatter(df['Size'], df['Rating'], alpha=0.5, color='r')
plt.xlabel('App Size')
plt.ylabel('Rating')
plt.title('App Size vs. Rating')
plt.grid(True)
plt.show()
```

## Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.

```python
# Convert the 'Price' column to numeric, replacing non-numeric values with NaN
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

# Calculate average prices for each app type
average_prices_by_type = df.groupby('Type')['Price'].mean()

# Create a bar chart to compare average prices by app type
plt.figure(figsize=(8, 6))
average_prices_by_type.plot(kind='bar', color='skyblue')
plt.xlabel('App Type')
plt.ylabel('Average Price')
plt.title('Average Prices by App Type')
plt.grid(axis='y')
plt.xticks(rotation=0)
plt.show()
```

## Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps with the highest number of installs.

```python
frequency_table = df['Installs'].value_counts()

# Sort the frequency table in descending order
sorted_frequency_table = frequency_table.sort_values(ascending=False)

# Display the top 10 most popular apps
top_10_popular_apps = sorted_frequency_table.head(10)

print(top_10_popular_apps)
```
